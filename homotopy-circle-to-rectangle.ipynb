# Cauchy's Theorem

**Definition.** Suppose $ \gamma_0 $ and  $ \gamma_1 $ are closed paths in the region G ⊆ C, parametrized by $ \gamma_0(t) $ , 0 ≤ t ≤ 1, and  $ \gamma_1(t)$ , 0 ≤ t ≤ 1, respectively. Then $ \gamma_0 $ is G-homotopic to $ \gamma_1$ if there exists a continuous
function h : $[ 0, 1 ]^2 \rightarrow G $ such that, for all s, t ∈ [ 0, 1 ], <br>
$ h(t, 0) = \gamma_0(t) $ , <br>
$ h(t, 1) =  \gamma_1(t) $, <br>
$ h(0, s) = h(1, s)$. <br>  <br>
The function $h(t,s)$ is called a homotopy [1].


### Homotopy of rectangle and ellipse 
Deformation of a rectangle into an ellipse.


```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.assumptions.assume import global_assumptions
init_printing()
```

We need to define all variables as real. Otherwise the funtions will not split into Re and Im part effectively.

```python
var('x, y, z, t, s', real=True)
```

The homotopy is <br>
$ h(t,s) = (1-s) e^{2 \pi i t} + 3s y(t) $ <br>
where $y(t)$ is the piecewise continuous function defined below. <br>

The curve is plotted with x = Re$(h(t,s))$ and y = Im$(h(t,s))$




A piecewise function y(t) divides the rectangle into 8 parts and helps map each part onto the ellipse.

```python
yt = [1 + 8* I * t,2-8*t+I,2-8*t+I,-1+4*I*(1-2*t),-1+4*I*(1-2*t),8*t-6-I,8*t-6-I,1 + 8* I * (t-1)]
yt
```

```python
for ss in np.arange(0,1,0.1):
    figs=[]
    i=0
    for y in yt:
        h = (1-s)*exp(2*pi*I*t) + 3*s*y
        h0 = h.subs({s:ss})
        a,b=h0.as_real_imag()

        x = [a.subs({t:tt}) for tt in np.linspace(i*1/8,i*1/8+1/8)]
        y = [b.subs({t:tt}) for tt in np.linspace(i*1/8,i*1/8+1/8)]

        p = plt.plot(x,y)
        figs.append(p[0])
        i=i+1
```

The inner loop generates each closed curve by looping over the piecewise function. 
The outer loop generates the homotopy (s:[0,1]).

This doesn't look as smooth as Fig. 4.1 of [1].
